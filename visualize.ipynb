```python
f = open("pos_ls_x.txt")
lines = f.readlines()
x_pos_ls=[]
for line in lines: 
    line=line.rstrip().split(",")
    line=[float(x) for x in line]
    x_pos_ls.append(line)
```

```python
f = open("pos_ls_y.txt")
lines = f.readlines()
y_pos_ls=[]
for line in lines: 
    line=line.rstrip().split(",")
    line=[float(x) for x in line]
    y_pos_ls.append(line)
```

```python
nx=500
dt=0.001
import matplotlib.pyplot as plt 
figure=plt.figure(figsize=(10,10))
ax = plt.subplot(1,2,1)
ax.scatter(0,-1,marker="x",s=200)
ax.scatter(0,1,marker="o",s=200)

for idx in range(len(y_pos_ls)):
    if len(x_pos_ls[idx]) >= 20000:
        pass
    else:
        ax.plot(x_pos_ls[idx], y_pos_ls[idx],label="trace")
```
